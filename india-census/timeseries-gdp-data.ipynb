# Manipulating GDP data

```python
# Function to append two gdp csv file
def appendGdp(state1,state2):
    state1 =  state1.rename(columns=lambda x: x.strip())
    state2 =  state2.rename(columns=lambda x: x.strip())
    state  =  state1.append(state2,sort=False)[state1.columns.tolist()].reset_index()
    state  =  state.drop(columns=['index'], axis = 0)
    return state

# Function to change column header of gdp dataframe
def changeGdp(stateName):
    stateName         = stateName.rename(columns=lambda x: x.strip())
    stateName         = stateName.transpose().reset_index()
    stateName.columns = stateName.iloc[1] + '|' + stateName.iloc[0]
    stateName         = stateName.rename(columns={'Description|Year':'District'})
    stateName         = stateName.drop([0, 1])
    return stateName

# Manupulating different states gdp data
AndhraPradesh = appendGdp(AndhraPradesh1 ,AndhraPradesh2)
AndhraPradesh = AndhraPradesh.drop([5,6,7,8,14,15,16])
AndhraPradesh = changeGdp(AndhraPradesh)
AndhraPradesh.insert(0,'State','ANDHRA PRADESH')

ArunachalPradesh = changeGdp(ArunachalPradesh1)
ArunachalPradesh.insert(0,'State','ARUNACHAL PRADESH')

Assam         = appendGdp(Assam1,Assam2)
Assam         = changeGdp(Assam)
Assam.insert(0,'State','ASSAM')

Bihar         = appendGdp(Bihar1,Bihar2)
Bihar         = Bihar.drop([5])
Bihar         = changeGdp(Bihar)
Bihar.insert(0,'State','BIHAR')

Chattisgarh = changeGdp(Chattisgarh1)
Chattisgarh.insert(0,'State','CHATTISGARH')


Haryana     =  changeGdp(Haryana1)
Haryana.insert(0,'State','HARYANA')

HimachalPradesh = changeGdp(HimachalPradesh1)
HimachalPradesh.insert(0,'State','HIMACHAL PRADESH')

Jharkhand    = changeGdp(Jharkhand1)
Jharkhand.insert(0,'State','JHARKHAND')

Karnataka   = appendGdp(Karnataka1,Karnataka2)
Karnataka   = changeGdp(Karnataka)
Karnataka.insert(0,'State','KARNATAKA')

Kerala      =  appendGdp(Kerala1,Kerala2)
Kerala      = Kerala.drop([6,7,14])
Kerala      = changeGdp(Kerala)
Kerala.insert(0,'State','KERALA')

MadhyaPradesh = changeGdp(MadhyaPradesh1)
MadhyaPradesh.insert(0,'State','MADHYA PRADESH')

Maharashtra  = appendGdp(Maharashtra1,Maharashtra2)
Maharashtra  = Maharashtra.drop([6,7,14])
Maharashtra  = changeGdp(Maharashtra)
Maharashtra.insert(0,'State','MAHARASHTRA')

Manipur      = changeGdp(Manipur1)
Manipur.insert(0,'State','MANIPUR')

Meghalaya    = changeGdp(Meghalaya1)
Meghalaya.insert(0,'State','MEGHALAYA')

Mizoram     = changeGdp(Mizoram1)
Mizoram.insert(0,'State','MIZORAM')

Odisha      = appendGdp(Odisha1,Odisha2)
Odisha      = Odisha.drop([5,6,7,13,14])
Odisha      = changeGdp(Odisha)
Odisha.insert(0,'State','ODISHA')

Punjab      = appendGdp(Punjab1,Punjab2)
Punjab      = Punjab.drop([5,6,12])
Punjab      = changeGdp(Punjab)
Punjab.insert(0,'State','PUNJAB')

Rajasthan   = appendGdp(Rajasthan1,Rajasthan2)
Rajasthan   = Rajasthan.drop([5,6,12])
Rajasthan   = changeGdp(Rajasthan)
Rajasthan.insert(0,'State','RAJASTHAN')

Sikkim      = changeGdp(Sikkim1)
Sikkim.insert(0,'State','SIKKIM')

Tamilnadu   = changeGdp(Tamilnadu1)
Tamilnadu.insert(0,'State','TAMIL NADU')

Uttarakhand = changeGdp(Uttarakhand1)
Uttarakhand.insert(0,'State','UTTARAKHAND')

UttarPradesh = appendGdp(UttarPradesh1,UttarPradesh2)
UttarPradesh = UttarPradesh.drop([5,6,12])
UttarPradesh = changeGdp(UttarPradesh)
UttarPradesh.insert(0,'State','UTTAR PRADESH')

WestBengal  = appendGdp(WestBengal1,WestBengal2)
WestBengal  = WestBengal.drop([5,6,7,13,14])
WestBengal  = changeGdp(WestBengal)
WestBengal.insert(0,'State','WEST BENGAL')

# Appending all the states into single dataframe
dataGDP = AndhraPradesh.append([ArunachalPradesh,Assam,Bihar,Chattisgarh,Haryana,HimachalPradesh,Jharkhand,Karnataka,
          Kerala,MadhyaPradesh,Maharashtra,Manipur,Meghalaya,Mizoram,Odisha,Punjab,Rajasthan,Sikkim,Tamilnadu,Uttarakhand,
          UttarPradesh,WestBengal],sort=False)
dataGDP['District'] = dataGDP['District'].str.title()
```

# Methods

```python
# Geo code and geo level
geography = pd.read_csv('data/geography.csv',encoding='UTF-8')
geo = geography[['geo_level','geo_code','name']]

# Calculating country data
def setCountryData(df,cols):
    for x in cols:
         df.loc[df['geo_level'] == 'country', x]= df.loc[df['geo_level'] == 'state', x].sum()
    return df;


# Calculating census data
def getCensusData(df,year):
    cols = list(df.columns.values)[3:]
    df.insert(2,'geo_version','2011')
    df.insert(3,'gdpyear',year)
    df = df.rename(index=str, columns={cols[0]: 'total'})
    df['total'] = df['total'].astype(int)
    df = df.sort_values(by=['gdpyear','geo_level','name']).reset_index().drop(['index'],axis=1).drop(['name'],axis=1)
    return  df

# Calculating geographic data
def getGdpData(df,year):
    cols = list(df.columns.values)[2:]
    df = df.replace(',','',regex=True)
    df[cols] = df[cols].astype(float)
    df = df.rename(index=str, columns={'State': 'name'})
    dfState = df.groupby(['name'])[cols].sum().reset_index()
   
    dfDis   = df.groupby(['District'])[cols].sum().reset_index()
    dfDis   = dfDis.rename(index=str, columns={'District': 'name'})
    dfDis['name'] = dfDis['name'].str.title()
    
    dfCom = pd.concat([dfState, dfDis])
    dfgeo = geo.merge(dfCom ,on=['name'],how='left')
    dfgeo = dfgeo.sort_values(by=['geo_level','name'])
    dfgeo = dfgeo.fillna(method = 'bfill', axis=0).fillna(0)
    dfgeoCountry = setCountryData(dfgeo,cols)
    gdp   = getCensusData(dfgeoCountry,year)
    return gdp
```
