```python
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.ticker import FuncFormatter
from IPython.display import display, display_pretty, Javascript, HTML
from pandas_highcharts.core import serialize
from pandas_highcharts.display import display_charts
```

```python
file1_path = "data/india_census_housing-hlpca-full.csv"
file2_path = "data/india-districts-census-2011.csv"
pd.set_option('display.float_format', lambda x: '%.3f' % x)
dataFirst = pd.read_csv(file1_path, encoding='utf-8')
dataSecond = pd.read_csv(file2_path, encoding='utf-8')
dataSecond['Literate Per'] = (dataSecond['Literate'].div(dataSecond['Population'])).multiply(100)
```

<h1> State Wise Population In 2001 </h1>

```python
dfp = dataSecond.groupby(['State Name'])[['Male','Female','Population']].sum().rename(columns={'Population':'Total'})
display_charts(dfp , kind="bar", title="Population",figsize = (1000, 700))
```

<h1> Caste Distribution In India </h1>

```python
df = dataSecond.groupby(['State Name'])[['SC','Male_SC','Female_SC','ST','Male_ST','Female_ST']].sum().rename(columns={'Male_SC':'SC Male','Female_SC':'SC Female','Male_ST':'ST Male','Female_ST':'ST Female'})
display_charts(df,kind="bar", title="Caste Distribution In India",figsize = (1000, 700))
```

<h1> Literacy In India </h1>

```python
df2 = dataSecond.groupby(['State Name'])[['Literate Per']].sum().div(dataSecond.groupby(['State Name'])[['Literate Per']].count()) 
display_charts(df2, kind='pie', title='Literacy In India', tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

<h1> Distribution Of Religions In India </h1>

```python
dfr = dataSecond[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains','Others_Religions']].sum()
res = pd.DataFrame(dfr[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains','Others_Religions']])
display_charts(res, kind='pie', title='Religions Distribution In India', tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

<h1> State Wise Literacy Percentage In India </h1>

```python
df2 = dataSecond.groupby(['State Name'])[['Literate Per']].sum().div(dataSecond.groupby(['State Name'])[['Literate Per']].count())
display_charts(df2,kind="bar", title="State Wise Literacy Percentage",figsize = (1000, 700))
```

<h1> State Wise Educations In India </h1>

```python
edu = dataSecond.groupby(['State Name'])[['Below_Primary_Education','Primary_Education','Middle_Education','Secondary_Education','Higher_Education','Graduate_Education','Other_Education','Literate','Illiterate_Education']].sum().rename(columns={'Below_Primary_Education':'Below Primary','Primary_Education':'Primary','Middle_Education':'Middle','Secondary_Education':'Secondary','Higher_Education':'Higher','Graduate_Education':'Graduate','Other_Education':'Other','Illiterate_Education':'Illiterate'})
display_charts(edu,title="State Wise Educations In India",figsize = (1000, 600))
```

<h1> Percentage Of Different Cooking Source In Different States Of India </h1>

```python
cs = dataFirst.groupby(['State Name'])[['Cooking_FW','Cooking_CR','Cooking_CC','Cooking_CLC','Cooking_kerosene','Cooking_LPG_PNG','Cooking_Electricity','Cooking_Biogas','Cooking_AO','Cooking_NC']].sum().div(dataFirst.groupby(['State Name'])[['Cooking_FW','Cooking_CR','Cooking_CC','Cooking_CLC','Cooking_kerosene','Cooking_LPG_PNG','Cooking_Electricity','Cooking_Biogas','Cooking_AO','Cooking_NC']].count()).rename(columns={'Cooking_FW':'Fire Wood','Cooking_CR':'Crop Residue','Cooking_CC':'Cowdung Cake','Cooking_CLC':'Charcoal','Cooking_kerosene':'Kerosene','Cooking_LPG_PNG':'LPG/PNG','Cooking_Electricity':'Electricity','Cooking_Biogas':'Biogas','Cooking_AO':'Other','Cooking_NC':'No Cooking'})
display_charts(cs,kind= 'bar', title="State Wise Different Cooking Source Percentage In India",figsize = (1000, 600))
```

<h1> Cooking Source Distribution In India </h1>

```python
tempdf = dataFirst.groupby(['State Name'])[['Cooking_FW','Cooking_CR','Cooking_CC','Cooking_CLC','Cooking_kerosene','Cooking_LPG_PNG','Cooking_Electricity','Cooking_Biogas','Cooking_AO','Cooking_NC']].sum().div(dataFirst.groupby(['State Name'])[['Cooking_FW','Cooking_CR','Cooking_CC','Cooking_CLC','Cooking_kerosene','Cooking_LPG_PNG','Cooking_Electricity','Cooking_Biogas','Cooking_AO','Cooking_NC']].count()).rename(columns={'Cooking_FW':'Fire Wood','Cooking_CR':'Crop Residue','Cooking_CC':'Cowdung Cake','Cooking_CLC':'Charcoal','Cooking_kerosene':'Kerosene','Cooking_LPG_PNG':'LPG/PNG','Cooking_Electricity':'Electricity','Cooking_Biogas':'Biogas','Cooking_AO':'Other','Cooking_NC':'No Cooking'}).reset_index()
csdf = pd.DataFrame(tempdf[['Fire Wood','Crop Residue','Cowdung Cake','Charcoal','Kerosene','LPG/PNG','Electricity','Biogas','Other','No Cooking']].sum())
display_charts(csdf, kind='pie', title='Cooking Source Distribution In India', tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

<h1> Main Source Of Drinking Water In India </h1>
