```python
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt

path_to_file = "data/india-districts-census-2011.csv"
data = pd.read_csv(path_to_file, encoding='utf-8')

#data.head()
data.groupby('State name').first()
```

```python
plot_data = data.groupby('State name')['Literate'].sum()
plot_data.plot(kind='bar',  stacked=True, title="Literacy data", figsize=(12, 7))
plt.ylabel('Literacy')
```

```python
data.groupby('State name')[['Male','Female']].first().head()
plot_data = data.groupby('State name')['Male','Female'].sum()
plot_data.plot(kind='bar', figsize=(15, 7)) #stacked=True
plt.ylabel('Population')
```

```python
data.groupby('State name')[['Male_Literate','Female_Literate']].first().head()
plot_data = data.groupby('State name')['Male_Literate','Female_Literate'].sum()
plot_data.plot(kind='bar', figsize=(12, 7)) #stacked=True
plt.ylabel('Literate ratio')
```
