```python
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt

path_to_file = "data/india-districts-census-2011.csv"
data = pd.read_csv(path_to_file, encoding='utf-8')

data.head()
```

```python
data.groupby('State name').first()
```

```python
plot_data = data.groupby('State name')['Literate'].sum()
plot_data.plot(kind='bar')
plt.ylabel('Literacy')
```
