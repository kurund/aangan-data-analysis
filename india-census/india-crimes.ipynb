```python
import pandas as pd
import numpy as np
from pandas_highcharts.core import serialize
from pandas_highcharts.display import display_charts
from IPython.display import display,HTML
pd.options.display.max_columns = None
from datascience import Table
from functools import reduce

# Reading csv data
dataDis2011      = Table.read_table('data/india-districts-census-2011.csv')
dataHous2011     = pd.read_csv('data/india_census_housing-hlpca-full.csv', encoding='utf-8')
dataMissChildren = Table.read_table('data/missing_traced_children.csv')
dataCrime        = Table.read_table('data/crimes.csv')
dataCyberCrm     = Table.read_table('data/cyber_crime.csv')
dataChldCrm      = Table.read_table('data/crime_against_children.csv')
dataTrafficing   = Table.read_table('data/trafficing.csv')
dataRapeVictm    = Table.read_table('data/rape_victims.csv')
dataKidnapping   = Table.read_table('data/kidnapping.csv')
dataJuvenileCrm  = Table.read_table('data/juvenile_crime.csv')
dataMurderVictm  = Table.read_table('data/murder_victims.csv')
dataCorruption   = Table.read_table('data/coruption_case.csv')
dataCurSeized    = Table.read_table('data/currency_siezed.csv')
dataJuvArrest    = Table.read_table('data/juvenile_arrested_bckgrnd.csv')
dataMrdrMotive   = Table.read_table('data/murder_motives.csv')
dataVioCrm       = Table.read_table('data/violent_crime_incidence.csv')
```

# Literacy and crime rate of different states of India

```python
dataCensus2011 = dataDis2011.relabeled(1,'STATE NAME').relabeled(2,'DISTRICT NAME')
# Manupulating literacy rate
censusLR       = dataCensus2011.select(1,'Population','Literate')
Literacy_rate  = (censusLR.column(2)/censusLR.column(1))*100
censusLRN      = censusLR.with_column('LITERACY RATE',Literacy_rate).drop(1,2)
censusLRdf     = censusLRN.to_df()
censusLRdf.replace('NCT OF DELHI','DELHI',inplace=True)
censusLRdfg    = censusLRdf.groupby(['STATE NAME'])[['LITERACY RATE']].mean().round(2).reset_index()

# Manupulating crime rate
Crime_rate    = dataCrime.to_df()
Crime_ratedfg = Crime_rate.groupby(['STATE NAME'])[['CRIME RATE']].sum().reset_index()

# Merging both
Litcrmdf  = censusLRdfg.merge(Crime_ratedfg,on=['STATE NAME'])
Litcrmdfg = Litcrmdf.groupby(['STATE NAME'])[['LITERACY RATE','CRIME RATE']].sum()
display_charts(Litcrmdfg , kind="bar", title="State wise Literacy rate vs Crime rate",figsize = (1000, 700))
```

# Merged crime data and socio economic data at state level

```python
# Manupulating district data
dataDis2011 = dataDis2011.to_df()
dataDis2011.columns = map(str.upper, dataDis2011.columns)
dataDis2011['DISTRICT NAME'] = dataDis2011['DISTRICT NAME'].str.upper()
dataDis2011.replace('NCT OF DELHI','DELHI',inplace=True)
dataDis2011.replace('ORISSA','ODISHA',inplace=True)
dataDisByState = dataDis2011.groupby(['STATE NAME'])[dataDis2011.columns.values].sum().reset_index()

# Merging socio data by crime data at state level
dataSocCrm = dataDisByState.merge(crimedfCom,on=['STATE NAME'])
dataSocCrm
```
