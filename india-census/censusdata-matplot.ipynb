# Combining two csv file to find household from first and literacy from second

```python
file1_path = "data/india_census_housing-hlpca-full.csv"
file2_path = "data/india-districts-census-2011.csv"

dataFirst = pd.read_csv(file1_path, encoding='utf-8')
dataSecond = pd.read_csv(file2_path, encoding='utf-8')
results = pd.merge(dataFirst[['State Name','Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated']],dataSecond[['State Name','Literate']],on='State Name')
results.groupby(['State Name'])[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated','Literate']].sum()
```

```python
results.groupby(['State Name'])[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated']].sum().plot.pie(subplots=True, figsize=(30, 5))
```

```python
rdf = results.groupby(['State Name'])[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated']].sum().reset_index()
fig, axes = plt.subplots(8, 3, figsize=(12, 9))

for i, (idx, row) in enumerate(rdf.set_index('State Name').iterrows()):
    ax = axes[i // 4, i % 3]
    row = row[row.gt(row.sum() * .01)]
    ax.pie(row, labels=row.index, startangle=30)
    ax.set_title(idx)

fig.subplots_adjust(wspace=.3)
```

```python
results.groupby(['State Name'])[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated']].sum().plot.bar(figsize=(12, 7))
```

```python
results.groupby(['State Name'])[['Literate']].sum().plot.pie(subplots=True, figsize=(8, 8))
```

```python
results.groupby(['State Name'])[['Literate']].sum().plot.bar(figsize=(12, 7), alpha=0.5)
```
