# India Primary Health Care

```python
import pandas as pd
import numpy as np
# Read health care csv file
AlloDocPHCS        = pd.read_csv('data/health/allo-doc-PHCS.csv', encoding='utf-8')
DisSubdisDoc       = pd.read_csv('data/health/dis-subdis-doctors.csv', encoding='utf-8')
PhysicianCHCS      = pd.read_csv('data/health/physician-CHCS.csv', encoding='utf-8')
SurgeonCHCS        = pd.read_csv('data/health/surgeons-CHCS.csv', encoding='utf-8')
FunctioningPHCS    = pd.read_csv('data/health/year-wise-PHCS.csv', encoding='utf-8')
FunctioningCHCS    = pd.read_csv('data/health/year-wise-CHCS.csv', encoding='utf-8')
FunctioningSubcen  = pd.read_csv('data/health/year-wise-Subcentre.csv', encoding='utf-8')
WorkerMaleSubCen   = pd.read_csv('data/health/worker-male-subcen.csv', encoding='utf-8')
WorkerFemaleSubCen = pd.read_csv('data/health/worker-female-subcen.csv', encoding='utf-8')
AssisMalePHCS      = pd.read_csv('data/health/assistant-male-PHCS.csv', encoding='utf-8')
AssisFemalePHCS    = pd.read_csv('data/health/assistant-female-PHCS.csv', encoding='utf-8')
NurseStaffPHCCHC   = pd.read_csv('data/health/nursing-staff-PHCS-CHCS.csv', encoding='utf-8')
FacilitiesCHCS     = pd.read_csv('data/health/facilities-CHCS.csv', encoding='utf-8')
FacilitiesPHCS     = pd.read_csv('data/health/facilities-PHCS.csv', encoding='utf-8')
```

# Allopathic doctors as PHCs

```python
AlloDocPHCSdf = AlloDocPHCS[['State/ UT','Required - [R]','In Position - [P]']]
AlloDocPHCSdf = AlloDocPHCSdf.rename(index=str, columns={'Required - [R]': 'Allopathic doctor_PHCs_Required','In Position - [P]':'Allopathic doctor_PHCs_Available'})
AlloDocPHCSdf.head()
```

# Dis and sub-dis doctors

```python
DisSubdisDocdf = DisSubdisDoc[['State/ UT','District Hospital - In Position','Sub District/ Sub Divisional Hospital - In Position']]
DisSubdisDocdf = DisSubdisDocdf.rename(index=str, columns={'District Hospital - In Position': 'Doctors_District_Hospital','Sub District/ Sub Divisional Hospital - In Position':'Doctors_Sub District_Hospital'})
DisSubdisDocdf.head()
```

# Physicians at CHCs

```python
PhysicianCHCSdf = PhysicianCHCS[['State/ UT','Required - [R]','In Position - [P]']]
PhysicianCHCSdf = PhysicianCHCSdf.rename(index=str, columns={'Required - [R]' :'Physician_Required_CHCS','In Position - [P]':'Physician_Available_CHCS'})
PhysicianCHCSdf.head()
```

# Surgeons CHCS

```python
SurgeonCHCSdf = SurgeonCHCS[['State/ UT','Required - [R]','In Position - [P]']]
SurgeonCHCSdf  = SurgeonCHCSdf .rename(index=str, columns={'Required - [R]' :'Surgeon_Required_CHCS','In Position - [P]':'Surgeon_Available_CHCS'})
SurgeonCHCSdf.head()
```

# Year wise PHCS functioning

```python
colsPHCS = list(FunctioningPHCS.columns.values)[1:]
FunctioningPHCSdf = FunctioningPHCS[colsPHCS]
FunctioningPHCSdf.columns = FunctioningPHCSdf.columns.str.split('[').str[1].str.split(']').str[0]
FunctioningPHCSdf.columns = ['PHCS_' + str(col) for col in FunctioningPHCSdf]
FunctioningPHCSdf = FunctioningPHCSdf.rename(index=str, columns={'PHCS_nan':'State/ UT'})
FunctioningPHCSdf.head()
```

# Year wise CHCS functioning

```python
colsCHCS = list(FunctioningCHCS.columns.values)[1:]
FunctioningCHCSdf =  FunctioningCHCS[colsCHCS]
FunctioningCHCSdf.columns = FunctioningCHCSdf.columns.str.split('[').str[1].str.split(']').str[0]
FunctioningCHCSdf.columns = ['CHCS_' + str(col) for col in FunctioningCHCSdf]
FunctioningCHCSdf = FunctioningCHCSdf.rename(index=str, columns={'CHCS_nan':'State/ UT'})
FunctioningCHCSdf.head()
```

# Year wise subcentre functioning

```python
colsSubC = list(FunctioningSubcen.columns.values)[1:]
FunctioningSubcendf =  FunctioningSubcen[colsSubC]
FunctioningSubcendf.columns = FunctioningSubcendf.columns.str.split('[').str[1].str.split(']').str[0]
FunctioningSubcendf.columns = ['SubCentre_' + str(col) for col in FunctioningSubcendf]
FunctioningSubcendf = FunctioningSubcendf.rename(index=str, columns={'SubCentre_nan':'State/ UT'})
FunctioningSubcendf.head()
```

# Workers at subcentre

```python
WorkerMaleSubCendf =  WorkerMaleSubCen[['State/ UT','Required - [R]','In Position - [P]']]
WorkerMaleSubCendf =  WorkerMaleSubCendf.rename(index=str, columns={'Required - [R]':'Workers_Required_SubCen_Male','In Position - [P]':'Workers_Available_SuBCen_Male'})
WorkerMaleSubCendf

WorkerFemaleSubCendf = WorkerFemaleSubCen[['State/ UT','Required - [R]','In Position - [P]']]
WorkerFemaleSubCendf = WorkerFemaleSubCendf.rename(index=str, columns={'Required - [R]':'Workers_Required_SubCen_Female','In Position - [P]':'Workers_Available_SuBCen_Female'})
WorkerFemaleSubCendf

WorkersSubCen = pd.merge(WorkerMaleSubCendf,WorkerFemaleSubCendf, on =['State/ UT'])
WorkersSubCen.head()
```

# Assistant as PHCS

```python
AssisMalePHCSdf = AssisMalePHCS[['State/ UT','Required - [R]','In Position - [P]']]
AssisMalePHCSdf =  AssisMalePHCSdf.rename(index=str, columns={'Required - [R]':'Assistant_Required_PHCS_Male','In Position - [P]':'Assistant_Available_PHCS_Male'})
AssisMalePHCSdf

AssisFemalePHCSdf =  AssisFemalePHCS[['State/ UT','Required - [R]','In Position - [P]']]
AssisFemalePHCSdf =  AssisFemalePHCSdf.rename(index=str, columns={'Required - [R]':'Assistant_Required_PHCS_Female','In Position - [P]':'Assistant_Available_PHCS_Female'})
AssisFemalePHCSdf

AssistantPHCS   = pd.merge(AssisMalePHCSdf,AssisFemalePHCSdf, on =['State/ UT'])
AssistantPHCS.head()
```

# Nursing staff at PHCS and CHCS

```python
NurseStaffPHCCHCdf = NurseStaffPHCCHC[['State/ UT','Required - [R]','In Position - [P]']]
NurseStaffPHCCHCdf = NurseStaffPHCCHCdf.rename(index=str, columns={'Required - [R]':'NurseStaff_Required_PHCS_CHCS','In Position - [P]':'NurseStaff_Available_PHCS_CHCS'})
NurseStaffPHCCHCdf.head()
```

# CHCS facilities

```python
colsFacCHC = list(FacilitiesCHCS.columns.values)
FacilitiesCHCSdf = FacilitiesCHCS[colsFacCHC]
FacilitiesCHCSdf.columns = FacilitiesCHCSdf.columns.str.split('-').str[1].str.strip()
FacilitiesCHCSdf  = FacilitiesCHCSdf.rename(index=str, columns={'With functional Laboratory':'Laboratory','With functional O.T.':'O.T','With functional Labor Room':'Labor Room','With New Born Care Corner':'New Born Care Corner','With functional X':'X-ray','With quarters for specialist Doctors':'Specialist Doctors Quarter','Allopathic drugs for common ailments':'Allopatic Drugs','AYUSH drugs for common ailments':'Ayush Drugs','With referral transport available':'Referral Transport','With at least 30 beds':'Min 30 beds'})
FacilitiesCHCSFdf = FacilitiesCHCSdf[['Col. 1','Min 30 beds','Laboratory','O.T','Labor Room','New Born Care Corner','X-ray','Specialist Doctors Quarter','Referral Transport','Allopatic Drugs','Ayush Drugs']]
FacilitiesCHCSFdf.columns = ['Facilities_CHCS_' + str(col) for col in FacilitiesCHCSFdf]
FacilitiesCHCSFdf = FacilitiesCHCSFdf.rename(index=str, columns={'Facilities_CHCS_Col. 1':'State/ UT'})
FacilitiesCHCSFdf.head()
```

# PHCS facilities

```python
colsFacPHC = list(FacilitiesPHCS.columns.values)
FacilitiesPHCSdf = FacilitiesPHCS[colsFacPHC]
FacilitiesPHCSdf.drop([col for col in FacilitiesPHCSdf.columns if '%' in col],axis=1,inplace=True)
FacilitiesPHCSdf.drop([col for col in FacilitiesPHCSdf.columns if 'Without' in col],axis=1,inplace=True)
FacilitiesPHCSdf.drop(['S. No.', 'Registered RKS','No. of PHCs Functioning as per IPHS norms','Number of PHCs Functioning','Number of Primary Health Centres - PHCs functoning on 24X7 basis - Number'],axis=1,inplace=True)
FacilitiesPHCSdf.columns = FacilitiesPHCSdf.columns.str.split('-').str[1].str.strip()
FacilitiesPHCSdf = FacilitiesPHCSdf.rename(index=str, columns={'With Labour Room':'Labour Room','With OT':'OT','With at least 4 beds':'Min 4 Beds','With Telephone':'Telephone','With Computer':'Computer'})
FacilitiesPHCSdf.columns = ['Facilities_PHCS_' + str(col) for col in FacilitiesPHCSdf]
FacilitiesPHCSdf = FacilitiesPHCSdf.rename(index=str, columns={'Facilities_PHCS_nan':'State/ UT'})
FacilitiesPHCSdf.head()
```
