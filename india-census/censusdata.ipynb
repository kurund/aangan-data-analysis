```python
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.ticker import FuncFormatter
from IPython.display import display, display_pretty, Javascript, HTML
from pandas_highcharts.core import serialize
from pandas_highcharts.display import display_charts
```

<h1> Combined Percentage Of  Livelihood and Literate In Different States In India </h1>

```python
file1_path = "data/india_census_housing-hlpca-full.csv"
file2_path = "data/india-districts-census-2011.csv"
pd.set_option('display.float_format', lambda x: '%.3f' % x)
dataFirst = pd.read_csv(file1_path, encoding='utf-8')
dataSecond = pd.read_csv(file2_path, encoding='utf-8')
dataSecond['Literate Per'] = (dataSecond['Literate'].div(dataSecond['Population'])).multiply(100)
results = pd.merge(dataFirst[['State Name','Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated']],dataSecond[['State Name','Literate Per']],on='State Name')
dfl = results.groupby(['State Name'])[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated','Literate Per']].sum().div(results.groupby(['State Name'])[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated','Literate Per']].count()).rename(columns={'Contition_T_Good':'Good','Contition_T_Livable':'Livable','Contition_T_Dilapidated':'Dilapidated','Literate Per':'Literate'})
display_charts(dfl , kind="bar", title="Livelihood",figsize = (1100, 700))
```

<h1> Combined Percentage Of Populations And HouseHold Member Size <h1>

```python
dataSecond['Male Per'] = (dataSecond['Male'].div(dataSecond['Population'])).multiply(100)
dataSecond['Female Per'] = (dataSecond['Female'].div(dataSecond['Population'])).multiply(100)
comdf = pd.merge(dataFirst[['State Name','H_size_1','H_size_2','H_size_3','H_size_4','H_size_5','H_size_6_8','H_size_9']],dataSecond[['State Name','Male Per','Female Per']],on='State Name')
phdf = comdf.groupby(['State Name'])[['H_size_1','H_size_2','H_size_3','H_size_4','H_size_5','H_size_6_8','H_size_9','Male Per','Female Per']].sum().div(comdf.groupby(['State Name'])[['H_size_1','H_size_2','H_size_3','H_size_4','H_size_5','H_size_6_8','H_size_9','Male Per','Female Per']].count()).rename(columns={'H_size_1':'HH Size1','H_size_2':'HH Size 2','H_size_3':'HH Size 3','H_size_4':'HH Size 4','H_size_5':'HH Size 5','H_size_6_8':'HH Size 6-8','H_size_9':'HH Size 9 & Above'})
display_charts(phdf ,kind="area", title="HouseHold Size And Population Percentage In India",figsize = (1100, 700))
```

<h1> Combined Percentage Of HouseHolds Type And Dwelling Rooms Quantity</h1>

```python
dataSecond['Rural HouseHolds'] = dataSecond['Rural_Households'].div(dataSecond['Population']).multiply(100)
dataSecond['Urban HouseHolds'] = dataSecond['Urban_Households'].div(dataSecond['Population']).multiply(100)
dataSecond['Total HouseHolds'] = dataSecond['Households'].div(dataSecond['Population']).multiply(100)
compdf = pd.merge(dataFirst[['State Name','Dwelling_R_NER','Dwelling_R_OR','Dwelling_R_TR','Dwelling_R_TH_R','Dwelling_R_FR','Dwelling_R_FI_R','Dwelling_R_SRA']],dataSecond[['State Name','Rural HouseHolds','Urban HouseHolds','Total HouseHolds']],on='State Name')
hdrdf = compdf.groupby(['State Name'])[['Dwelling_R_NER','Dwelling_R_OR','Dwelling_R_TR','Dwelling_R_TH_R','Dwelling_R_FR','Dwelling_R_FI_R','Dwelling_R_SRA','Rural HouseHolds','Urban HouseHolds','Total HouseHolds']].sum().div(compdf.groupby(['State Name'])[['Dwelling_R_NER','Dwelling_R_OR','Dwelling_R_TR','Dwelling_R_TH_R','Dwelling_R_FR','Dwelling_R_FI_R','Dwelling_R_SRA','Rural HouseHolds','Urban HouseHolds','Total HouseHolds']].count()).rename(columns={'Dwelling_R_NER':'No Exclusive Room','Dwelling_R_OR':'One Room','Dwelling_R_TR':'Two Rooms','Dwelling_R_TH_R':'Three Rooms','Dwelling_R_FR':'Four Rooms','Dwelling_R_FI_R':'Five Rooms','Dwelling_R_SRA':'Six Rooms & Above'})
display_charts(hdrdf,kind="bar" ,title="Percentage Of Households And Dwelling Rooms In India",figsize = (1100, 700))
```

<h1> Combined Percentage Of Male Female and Married Couple </h1>

```python
dataSecond['Male Per'] = dataSecond['Male'].div(dataSecond['Population']).multiply(100)
dataSecond['Female Per'] = dataSecond['Female'].div(dataSecond['Population']).multiply(100)
comfm = pd.merge(dataFirst[['State Name','Married_C_N','Married_C_1','Married_C_2','Married_C_3','Married_C_4','Married_C_5']],dataSecond[['State Name','Male Per','Female Per']],on='State Name')
mfmdf =  comfm.groupby(['State Name'])[['Married_C_N','Married_C_1','Married_C_2','Married_C_3','Married_C_4','Married_C_5','Male Per','Female Per']].sum().div(comfm.groupby(['State Name'])[['Married_C_N','Married_C_1','Married_C_2','Married_C_3','Married_C_4','Married_C_5','Male Per','Female Per']].count()).rename(columns={'Married_C_N':'Married Couple None','Married_C_1':'Married Couple 1','Married_C_2':'Married Couple 2','Married_C_3':'Married Couple 3','Married_C_4':'Married Couple 4','Married_C_5':'Married Couple 5 & Above'})
display_charts(mfmdf , kind="bar" ,title="Percentage Of Male Female and Married Couple In India",figsize = (1100, 700))
```

<h1> Comibined Percentage Of HouseHold With Electric Lighting And Main Source Of Lighting </h1>

```python
dataSecond['Electric Lighting'] = dataSecond['Housholds_with_Electric_Lighting'].div(dataSecond['Households']).multiply(100)
comhwel = pd.merge(dataFirst[['State Name','MSL_Electricty','MSL_Kerosene','MSL_SE','MSL_OO','MSL_AO','MSL_NL']],dataSecond[['State Name','Electric Lighting']],on='State Name')
hweldf = comhwel.groupby(['State Name'])[['MSL_Electricty','MSL_Kerosene','MSL_SE','MSL_OO','MSL_AO','MSL_NL','Electric Lighting']].sum().div(comhwel.groupby(['State Name'])[['MSL_Electricty','MSL_Kerosene','MSL_SE','MSL_OO','MSL_AO','MSL_NL','Electric Lighting']].count()).rename(columns={'MSL_Electricty':'Electricity','MSL_Kerosene':'Kerosene','MSL_SE':'Solar Energy','MSL_OO':'Other Oil','MSL_AO':'Any Other','MSL_NL':'No Lighting'})
display_charts(hweldf , kind="bar" ,title="Percentage Of HouseHold With Electricity And Main Source Of Lighting",figsize = (1100, 700))
```
