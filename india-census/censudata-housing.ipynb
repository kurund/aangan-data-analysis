```python
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

path_to_file = "data/india_census_housing-hlpca-full.csv"
data = pd.read_csv(path_to_file, encoding='utf-8')

data.groupby('State Name')[['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated']].first().head()
```

```python
plot_data = data.groupby('State Name')['Contition_T_Good','Contition_T_Livable','Contition_T_Dilapidated'].sum()
plot_data.plot(kind='bar', figsize=(12, 7)) #stacked=True
plt.ylabel('Contition_T_Good')
```

<b>Displyaing household conditions<b/>
    #Total Number of households , Total Number of Good, Total Number of Livable, Total Number of Dilapidated

```python
pie_stname = data.groupby('State Name').agg('count')
pie_hh_good = data.groupby('Contition_T_Good').agg('count')
pie_hh_livable = data.groupby('Contition_T_Livable').agg('count')
pie_hh_Dilapidated = data.groupby('Contition_T_Dilapidated').agg('count')

pie_hh_good_labels = pie_stname.Contition_T_Good.sort_values().index
pie_hh_good_counts = pie_stname.Contition_T_Good.sort_values()

pie_hh_livable_labels = pie_stname.Contition_T_Livable.sort_values().index
pie_hh_livable_counts = pie_stname.Contition_T_Livable.sort_values()

pie_hh_Dilapidated_labels = pie_stname.Contition_T_Dilapidated.sort_values().index
pie_hh_Dilapidated_counts = pie_stname.Contition_T_Dilapidated.sort_values()


# Make square figures and axes
plt.figure(1, figsize=(20,10))
the_grid = GridSpec(2, 3)

cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]


plt.subplot(the_grid[0, 0], aspect=1, title='Total households with Good condition')
source_pie = plt.pie(pie_hh_good_counts, labels=pie_hh_good_labels, autopct='%1.1f%%', shadow=True, colors=colors)

plt.subplot(the_grid[0, 1], aspect=1, title='Total households with Livable condition')
flavor_pie = plt.pie(pie_hh_livable_counts,labels=pie_hh_livable_labels, autopct='%.0f%%', shadow=True, colors=colors)


plt.subplot(the_grid[0, 2], aspect=1, title='Total households with Dilapidated condition')
flavor_pie = plt.pie(pie_hh_Dilapidated_counts,labels=pie_hh_Dilapidated_labels, autopct='%.0f%%', shadow=True, colors=colors)

plt.suptitle('Total Number of households', fontsize=15)


plt.show()
```
