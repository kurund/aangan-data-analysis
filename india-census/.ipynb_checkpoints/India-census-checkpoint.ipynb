```python
import pandas as pd
import numpy as np
from pandas_highcharts.core import serialize
from pandas_highcharts.display import display_charts
from IPython.display import display,HTML
pd.options.display.max_columns = None

# Reading csv data
dataCensus_2001  = pd.read_csv('data/india_census_2001.csv', encoding='utf-8')
dataCensus_2011  = pd.read_csv('data/india_census_2011.csv', encoding='utf-8')
dataEl_2005_06 = pd.read_csv('data/elementary_2005_06.csv', encoding='utf-8')
dataEl_2006_07 = pd.read_csv('data/elementary_2006_07.csv', encoding='utf-8')
dataEl_2007_08 = pd.read_csv('data/elementary_2007_08.csv', encoding='utf-8')
dataEl_2008_09 = pd.read_csv('data/elementary_2008_09.csv', encoding='utf-8')
dataEl_2009_10 = pd.read_csv('data/elementary_2009_10.csv', encoding='utf-8')
dataEl_2010_11 = pd.read_csv('data/elementary_2010_11.csv', encoding='utf-8')
dataEl_2011_12 = pd.read_csv('data/elementary_2011_12.csv', encoding='utf-8')

dataCensus_2001['YEAR'] = dataCensus_2001['YEAR'].astype('str')
dataCensus_2011['YEAR'] = dataCensus_2011['YEAR'].astype('str')
```

# Socio data 2011

```python
dataCensus_2011
```

```python
dataPop_2001 = dataCensus_2001.groupby(['YEAR'])[['MALES','FEMALES','POPULATION']].sum().rename(columns={'POPULATION':'TOTAL','MALES':'MALE','FEMALES':'FEMALE'})
dataPop_2011 = dataCensus_2011.groupby(['YEAR'])[['MALE','FEMALE','POPULATION']].sum().rename(columns={'POPULATION':'TOTAL'})
dataPop = dataPop_2001.append(dataPop_2011)
display_charts(dataPop, title="India population", kind="bar",figsize = (1000, 700))
```

# Reading GDP data

```python
dfGDP = pd.read_csv('data/india_gdp.csv', index_col=0, na_values=['(NA)']).fillna(0)
gdpTable  = pd.read_csv('data/india_gdp.csv',encoding='utf-8',index_col=0, na_values=['(NA)']).fillna('NA')
gdpTable
```

# India's Yearly GDP

```python
dfGDP['GDP (in Rs. Cr.)|2004-05'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2004-05'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2005-06'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2005-06'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2006-07'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2006-07'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2007-08'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2007-08'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2008-09'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2008-09'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2009-10'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2009-10'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2010-11'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2010-11'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2011-12'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2011-12'],errors='coerce')

in_gdp = dfGDP[['GDP (in Rs. Cr.)|2004-05','GDP (in Rs. Cr.)|2005-06','GDP (in Rs. Cr.)|2006-07','GDP (in Rs. Cr.)|2007-08','GDP (in Rs. Cr.)|2008-09','GDP (in Rs. Cr.)|2009-10','GDP (in Rs. Cr.)|2010-11','GDP (in Rs. Cr.)|2011-12','GDP (in Rs. Cr.)|2012-13']].rename(columns={'GDP (in Rs. Cr.)|2004-05':'2004-05','GDP (in Rs. Cr.)|2005-06':'2005-06','GDP (in Rs. Cr.)|2006-07':'2006-07','GDP (in Rs. Cr.)|2007-08':'2007-08','GDP (in Rs. Cr.)|2008-09':'2008-09','GDP (in Rs. Cr.)|2009-10':'2009-10','GDP (in Rs. Cr.)|2010-11':'2010-11','GDP (in Rs. Cr.)|2011-12':'2011-12','GDP (in Rs. Cr.)|2012-13':'2012-13'}).sum()
in_gdp = pd.DataFrame(in_gdp.round(2))
in_gdp = in_gdp.rename(columns={0:'GDP in Rs (Cr)'})
display_charts(in_gdp, title="India's GDP", kind="bar",figsize = (1000, 700))
```

# India GDP growth rate

```python
dfGDP['Growth Rate % (YoY)|2008-09'] = pd.to_numeric(dfGDP['Growth Rate % (YoY)|2008-09'],errors='coerce')
in_gdpGrowth = dfGDP[['Growth Rate % (YoY)|2005-06','Growth Rate % (YoY)|2006-07','Growth Rate % (YoY)|2007-08','Growth Rate % (YoY)|2008-09','Growth Rate % (YoY)|2009-10','Growth Rate % (YoY)|2010-11','Growth Rate % (YoY)|2011-12','Growth Rate % (YoY)|2012-13']].mean()
in_gdpGrowth = pd.DataFrame(in_gdpGrowth.round(2))
in_gdpGrowth = in_gdpGrowth.rename(columns={0:'Growth Rate'})
display_charts(in_gdpGrowth, title="India's growth rate", kind="bar",figsize = (1000, 700))
```

# Student enrolment in gov and pvt school for different year

```python
# Appending elementary data from 2005-2012
dataEl = dataEl_2005_06 .append([dataEl_2006_07 , dataEl_2007_08, dataEl_2008_09,dataEl_2009_10,dataEl_2010_11,dataEl_2011_12])
dataEl['ENROLMENT GOV'] = dataEl['PRIMARY (ENR GOVT1)'] + dataEl['PRIMARY WITH UPPER PRIMARY (ENR GOVT2)'] + dataEl['PRIMARY WITH UPPER PRIMARY SEC/HIGHER SEC. (ENR GOVT3)']+ dataEl['UPPER PRIMARY ONLY (ENR GOVT4)']+ dataEl['UPPER PRIMARY WITH SEC./HIGHER SEC. (ENR GOVT5)']
dataEl['ENROLMENT PVT'] = dataEl['PRIMARY (ENR PVT1)'] + dataEl['PRIMARY WITH UPPER PRIMARY (ENR PVT2)'] + dataEl['PRIMARY WITH UPPER PRIMARY SEC/HIGHER SEC. (ENR PVT3)'] + dataEl['UPPER PRIMARY ONLY (ENR PVT4)'] + dataEl['UPPER PRIMARY WITH SEC./HIGHER SEC. (ENR PVT5)']
datEnrGovSc = dataEl.groupby(['YEAR'])[['ENROLMENT GOV','ENROLMENT PVT']].sum()
display_charts(datEnrGovSc, title="Student's enrolment for differenr year", kind="bar",figsize = (1000, 700))
```
