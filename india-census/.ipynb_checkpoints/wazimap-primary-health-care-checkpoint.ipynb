# Primary health functions

```python
# Geo code and geo level
geography = pd.read_csv('data/geography.csv',encoding='UTF-8')
geo = geography[['geo_level','geo_code','name']]

# Calculating country data
def setCountryData(df,cols):
    for x in cols:
         df.loc[df['geo_level'] == 'country', x]= df.loc[df['geo_level'] == 'state', x].sum()
    return df;

def getHealthGeo(df):
    dfgeo = geo.merge(df ,on=['name'],how='left')
    cols = list(df.columns.values)[1:]
    dfgeoCountry = setCountryData(dfgeo,cols)
    return dfgeoCountry;

def getHealthCensus(df,varname):
    dfgeo = getHealthGeo(df)
    dfgeo = dfgeo.fillna(0)
    dfgeoFinal = pd.melt(dfgeo, id_vars=['geo_level', 'geo_code','name'], var_name=varname, value_name="total")
    dfgeoFinal = dfgeoFinal.sort_values(by=['geo_level','name']).reset_index().drop(['index'],axis=1).drop(['name'],axis=1)
    dfgeoFinal['total'] = dfgeoFinal['total'].astype(int)
    return dfgeoFinal
```

# District wise health care centres

```python
def getDisGeoData(df):
    cols = list(df.columns.values)[2:]
    df = df.replace(',','',regex=True)
    df[cols] = df[cols].astype(float)
    dfState = df.groupby(['name'])[cols].sum().reset_index()
   
    dfDis   = df.groupby(['DW_District'])[cols].sum().reset_index()
    dfDis   = dfDis.rename(index=str, columns={'DW_District': 'name'})
    dfDis['name'] = dfDis['name'].str.title()
    
    dfCom = pd.concat([dfState, dfDis])
    dfgeo = geo.merge(dfCom ,on=['name'],how='left')
    dfgeo = dfgeo.sort_values(by=['geo_level','name'])
    dfgeo = dfgeo.fillna(method = 'bfill', axis=0).fillna(0)
    dfgeoCountry = setCountryData(dfgeo,cols)
    return dfgeoCountry

DWHealthCareCenGeo = getDisGeoData(DWHealthCareCendf)
DWHealthCareCenFinal = pd.melt(DWHealthCareCenGeo, id_vars=['geo_level', 'geo_code','name'], var_name='healthcares', value_name="total")
DWHealthCareCenFinal = DWHealthCareCenFinal.sort_values(by=['geo_level','name']).reset_index().drop(['index'],axis=1).drop(['name'],axis=1)
DWHealthCareCenFinal['total'] = DWHealthCareCenFinal['total'].astype(int)
DWHealthCareCenFinal['healthcares'] = DWHealthCareCenFinal['healthcares'].str.split('_').str[1]
DWHealthCareCenFinal

#DWHealthCareCenFinal.to_csv('data/healthcarecentre.csv', index=False)
```
