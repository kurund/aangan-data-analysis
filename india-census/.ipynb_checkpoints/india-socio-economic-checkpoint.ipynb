```python
import pandas as pd
from pandas_highcharts.core import serialize
from pandas_highcharts.display import display_charts
from IPython.display import display
pd.options.display.max_columns = None
import numpy as np

# Reading csv data
dfGDP         = pd.read_csv('data/india_gdp.csv', index_col=0, na_values=['(NA)']).fillna(0)
dataHous2011  = pd.read_csv('data/india_census_housing-hlpca-full.csv', encoding='utf-8')
dataDis2011   = pd.read_csv('data/india-districts-census-2011.csv', encoding='utf-8')
dataDis2011.columns = map(str.upper, dataDis2011.columns)
dataDis2011['DISTRICT NAME'] = dataDis2011['DISTRICT NAME'].str.upper()
dataDis2011.replace('NCT OF DELHI','DELHI',inplace=True)
dataDis2011.replace('ORISSA','ODISHA',inplace=True)
```

# State wise population in India

```python
SWPIIdfg =  dataDis2011.groupby(['STATE NAME'])[['MALE','FEMALE','POPULATION']].sum().rename(columns={'POPULATION':'TOTAL'})
display_charts(SWPIIdfg, title="India's population", kind="bar",figsize = (1000, 700))
```

# State wise literacy rate in India

```python
dataDis2011['LITERACY RATE'] = dataDis2011['LITERATE']/dataDis2011['POPULATION']*100
SWLRIIdfg = dataDis2011.groupby(['STATE NAME'])[['LITERACY RATE']].mean().round(2)
display_charts(SWLRIIdfg, title="State wise litearcy rate in India", kind="bar",figsize = (1000, 700))
```

# Religion's distribution in India

```python
IRPdf  = dataDis2011[['HINDUS','MUSLIMS','CHRISTIANS','SIKHS','BUDDHISTS','JAINS','OTHERS_RELIGIONS']].sum()
IRPdfg = pd.DataFrame(IRPdf)
IRPdfg = IRPdfg.rename(columns={0:'RELIGIONS'})
display_charts(IRPdfg, kind='pie', title="Religions in India", tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

# State wise SC populations percentage in India

```python
SWSPIIdfg = dataDis2011.groupby(['STATE NAME'])[['SC']].sum()
display_charts(SWSPIIdfg, kind='pie', title="State wise SC populations in India", tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

# State wise ST populations in India

```python
SWSTPIdf = dataDis2011.groupby(['STATE NAME'])[['ST']].sum()
display_charts(SWSTPIdf, kind='pie', title="State wise ST populations in India", tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

# Population of different types of age groups in India 

```python
SWDAGI = dataDis2011.groupby(['STATE NAME'])[['AGE_GROUP_0_29','AGE_GROUP_30_49','AGE_GROUP_50']].sum().rename(columns={'AGE_GROUP_0_29':'0-29 YEARS','AGE_GROUP_30_49':'30-49 YEARS','AGE_GROUP_50':'50 YEARS & ABOVE'})
display_charts(SWDAGI, title="State wise different age groups in India", kind="barh",figsize = (1000, 700))
```

# Different age  group percentage in India

```python
AGPI   = dataDis2011[['AGE_GROUP_0_29','AGE_GROUP_30_49','AGE_GROUP_50']].rename(columns={'AGE_GROUP_0_29':'0-29 YEARS','AGE_GROUP_30_49':'30-49 YEARS','AGE_GROUP_50':'50 YEARS & ABOVE'}).sum()
AGPIdf = pd.DataFrame(AGPI)
AGPIdf = AGPIdf.rename(columns={0:'AGE'})
display_charts(AGPIdf, kind='pie', title="Different age groups percentage in India", tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```

# India socio data

```python
# Combining housing and district census 2011
dataHous2011 = dataHous2011.replace('&','AND',regex=True)
dataHous2011.replace('NCT OF DELHI','DELHI',inplace=True)
dataHous2011.replace('PUDUCHERRY','PONDICHERRY',inplace=True)
dataHous2011.columns = map(str.upper, dataHous2011.columns)
dataHous2011['DISTRICT NAME'] = dataHous2011['DISTRICT NAME'].str.upper()
dataHous2011 = dataHous2011.loc[dataHous2011['RURAL/URBAN'] == 'Total']
dataCensus2011 = dataDis2011.merge(dataHous2011,on=['STATE NAME','DISTRICT NAME'])
dataCensus2011['STATE NAME'] = dataCensus2011['STATE NAME'].str.strip()
dataCensus2011
```

# India gdp data

```python
dataGDP  = pd.read_csv('data/india_gdp.csv',encoding='utf-8',index_col=0, na_values=['(NA)']).fillna('NA')
dataGDP.replace('24-PARGANAS (NORTH)','NORTH TWENTY FOUR PARGANAS',inplace=True)
dataGDP.replace('24-PARGANAS (SOUTH)','SOUTH TWENTY FOUR PARGANAS',inplace=True)
dataGDP['STATE NAME'] = dataGDP['STATE NAME'].str.strip()
dataGDP.replace('ORRISA','ODISHA',inplace=True)
dataGDP
```

# Combined gdp and socio data

```python
dataSocioC = dataCensus2011.merge(dataGDP,on=['STATE NAME','DISTRICT NAME'],how='left')
dataSocioC.replace(np.NaN,'NA',inplace=True)
dataSocioC
```

# India's yearly GDP

```python
dfGDP['GDP (in Rs. Cr.)|2004-05'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2004-05'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2005-06'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2005-06'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2006-07'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2006-07'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2007-08'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2007-08'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2008-09'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2008-09'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2009-10'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2009-10'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2010-11'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2010-11'],errors='coerce')
dfGDP['GDP (in Rs. Cr.)|2011-12'] = pd.to_numeric(dfGDP['GDP (in Rs. Cr.)|2011-12'],errors='coerce')

in_gdp = dfGDP[['GDP (in Rs. Cr.)|2004-05','GDP (in Rs. Cr.)|2005-06','GDP (in Rs. Cr.)|2006-07','GDP (in Rs. Cr.)|2007-08','GDP (in Rs. Cr.)|2008-09','GDP (in Rs. Cr.)|2009-10','GDP (in Rs. Cr.)|2010-11','GDP (in Rs. Cr.)|2011-12','GDP (in Rs. Cr.)|2012-13']].rename(columns={'GDP (in Rs. Cr.)|2004-05':'2004-05','GDP (in Rs. Cr.)|2005-06':'2005-06','GDP (in Rs. Cr.)|2006-07':'2006-07','GDP (in Rs. Cr.)|2007-08':'2007-08','GDP (in Rs. Cr.)|2008-09':'2008-09','GDP (in Rs. Cr.)|2009-10':'2009-10','GDP (in Rs. Cr.)|2010-11':'2010-11','GDP (in Rs. Cr.)|2011-12':'2011-12','GDP (in Rs. Cr.)|2012-13':'2012-13'}).sum()
in_gdp = pd.DataFrame(in_gdp.round(2))
in_gdp = in_gdp.rename(columns={0:'GDP in Rs (Cr)'})
display_charts(in_gdp, title="India's GDP", kind="bar",figsize = (1000, 700))
```

# India's gdp growth rate

```python
dfGDP['Growth Rate % (YoY)|2008-09'] = pd.to_numeric(dfGDP['Growth Rate % (YoY)|2008-09'],errors='coerce')
in_gdpGrowth = dfGDP[['Growth Rate % (YoY)|2005-06','Growth Rate % (YoY)|2006-07','Growth Rate % (YoY)|2007-08','Growth Rate % (YoY)|2008-09','Growth Rate % (YoY)|2009-10','Growth Rate % (YoY)|2010-11','Growth Rate % (YoY)|2011-12','Growth Rate % (YoY)|2012-13']].mean()
in_gdpGrowth = pd.DataFrame(in_gdpGrowth.round(2))
in_gdpGrowth = in_gdpGrowth.rename(columns={0:'Growth Rate'})
display_charts(in_gdpGrowth, title="India's gdp growth rate", kind="bar",figsize = (1000, 700))
```

# State wise percapita income of  India

```python
dataSocioC = dataSocioC.replace(',','', regex=True)
dataSocioC.replace('NA',0, inplace=True)
dataSocioC['GDP'] = ((dataSocioC['GDP (in Rs. Cr.)|2004-05'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2005-06'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2006-07'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2007-08'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2008-09'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2009-10'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2010-11'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2011-12'].astype(float) + dataSocioC['GDP (in Rs. Cr.)|2012-13'].astype(float))/9).round(2)
dataSocioC['PERCAPITA INCOME'] = ((dataSocioC['GDP']/dataSocioC['POPULATION'])*10000000).round(2)
SWPCIdfg = dataSocioC.groupby(['STATE NAME'])[['PERCAPITA INCOME']].sum().round(2)
display_charts(SWPCIdfg, title="State wise percapita income of India", kind="bar",figsize = (1000, 700))
```

# State wise percentage share of BPL population in India

```python
dataSocioC.loc[((dataSocioC['PERCAPITA INCOME']>1) & (dataSocioC['PERCAPITA INCOME']<27000)) , 'BPL POPULATION'] = dataSocioC['POPULATION']
SWPSBPdfg = dataSocioC.groupby(['STATE NAME'])[['BPL POPULATION']].sum()
display_charts(SWPSBPdfg, kind='pie', title="BPL populations in different states of India", tooltip={'pointFormat': '{series.name}: <b>{point.percentage:.1f}%</b>'})
```
